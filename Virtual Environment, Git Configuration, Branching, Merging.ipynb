# VİRTUAL ENVİRONMENT?

* Sanal ortamların oluşturulma nedeni üzerinde çalışılan birden çok projenin bir arada iken getirmiş olduğu dezavantajları (versiyon uyuşmazlıkları, kütüphane ve paket farklılıkları vb.) ortadan kaldırmak için bağımsız ve izole bir ortam oluşturulmasını sağlamaktır.


* Uyuşmazlıkların ve çakışmaların yanı sıra farklı sanal ortam kullanımlarının üzerinde uğraşılan projelerde bağımsızlık ve izole bir ortam sağladığı için herhangi bir veri kaybı esnasında projenin genelini değil sadece bulunduğu ortamın zarar görmesini sağlar. Aynı şekilde istemli bir silinme durumunda da sadece istenilen paketlerin silinmesine yardımcı olur.


* Örnek verecek olursak; Farklı Web projelerinde Django v3.0.7 ve v3.1 paketlerini kullandığımızı düşünelim sistem bu farklı paketleri ayırt edemediği ve bir arada kullanamayacağı için 2 pakette ortak olarak aynı kimlikte aynı dizinde yer almaktadır. Bu python ve projenin ilerleyişi için sorun olacağından her proje için sanal ortam oluşturmak kullanımı ve ilerleyişi kolaylaştıracağı gibi olası kayıplarında önüne geçmektedir.


## PYTHON VIRTUAL ENVIRONMENT

* İlk olarak virtual environment oluşturmak için bunu sağlayan Python Pip (Pip installs packages)’in kurulumunu ve versiyonunu kontrol edelim. Python eğer yüklü ise pip kuruludur. Değil ise aşağıdaki komut ile kurulumu tamamlayabiliriz.
## sudo apt install python3-pip  

* Pip Versiyonunu kontrol edelim.
## pip --version

* Projenizde seçtiğiniz bir Python sürümünü kullanmak isterseniz;
## virtualenv -p / usr / bin / python3 env_adı

* Şimdi pip kullanımı için virtual environment kurulumu yapalım.
## pip install virtualenv

* Pip virtual environment oluşturalım:
## python –m (Env_adı)

* Şimdi oluşturulan environmentımızı aktif edelim.
## source env_adı/bin/activate

* Windows için;
##  env_adı\Scripts\activate.bat

* İnaktif etmek için;
## source env_adı/bin/deactivate

* Virtual Environmentımıza gerekli paketleri (Pandas, Numpy, Pytz vb.) kurmak için:
## pip install paket_adı

* Paket kurulumunda .txt dosyanızın içine aldığınız paket isimlerini toplu olarak yüklenmesini isterseniz;
## pip install -r dosya_adınız.txt

* Environmentımızın içine kurulumunu tamamladığımız paketleri görmek istersek;
## pip freze

* Environmentımızda kullandığımız paketlerin listesini çıkarmak istersek;
## pip freze > dosya_adı.txt

* Environmentımızda herhangi istediğimiz bir paketi kaldırmak istersek;
## pip unistall paket_adı

   * Komutlarını kullanabiliriz. Daha fazla Pip komutu için ***Cheatsheet***

# CONDA VIRTUAL ENVIRONMENT

* İlk olarak aşağıdaki bağlantıdan Anaconda'nın sistemimize kurulumunu sağlayalım.

## Anaconda3

* Anaconda'nın kurulumunu tamamladınız ve conda komutunu kullanamıyorsanız ve istenilen python, ipython sürümlerini eklemede sorun yaşıyorsanız, PATH ortamınızla alakalı bir sorun olabilir. Bunun için condayı PATH'E eklememiz gerekir.
## export PATH=~/anaconda3/bin:$PATH

* sonrasında  ***conda –version*** ile kontrolünü sağlayabiliriz.

* Conda environmentımızı oluşturalım.
## conda create –n env_adı

* Oluşturduğumuz environment belli paketler içeriyor (örnek: python2.7) eğer paket veya sürüm değişikliği yapmak isterseniz;
***conda create –n env_adı keras=1.0.8*** (İstenilen sürümü yazabilirsiniz)

* Güncelleme yapılan environment içindeki paketlerde oluşan değişiklikleri görmek ve değişimleri eski durumuna döndürmek istersek;
## conda list --revisions
## conda install --revision 2

* Conda environment içindeki yüklü paketlerin listesini çıkarmak istersek;
## conda list –explicit > dosya_isminiz.txt

* Oluşturduğumuz conda environmentımızın içindeki yüklü paketleri görmek istersek;
## conda list

* Oluşturduğumuz Conda environmentını aktif ve inaktif etmek istersek;
## conda activate env_adı || conda deactivate

* Oluşturduğumuz environmentı yada içerisindeki paketleri silmek istersek;
## conda env remove –env_isminiz

* Ek olarak; Conda sürümünü güncellemek istersek;
## conda update conda

* Environment içine dosya ekleme ve mevcut dosyanın güncellemesini yapmak istersek;
## conda install paket_adı 
## conda update paket_adı

* Komutları ile environment içerisinde işlem yapabiliriz. 
* Daha fazla Conda komutu için Cheatsheet

# JUPYTER LAB
## Neden JUPYTER Lab?

* Popüler dosya formatlarını (Markdown, JSON, CSV, Vega) canlı ön izlemeler ile düzenleme şansının yanı sıra kullanımı basit ve birçok işlemi sekme kalabalığı yapmadan aynı proje içinde sistem ile yüksek etkileşimli bir yapıda gerçekleştirme şansı sunmasıdır.

* Jupyter Lab’i Conda ile yüklemek istediğimizde;
## conda install –c conda-forge jupyterlab

* Pip ile yüklemek istediğimizde;
## pip install jupyterlab 

* Aktif edilen Conda environmentımızın içinde iken environmentımızı kernel olarak Jupyter Lab’e  eklememiz gerekiyor. Kernel jupyter lab ile bağlantı sağlamaktadır. Bunu yapabilmek için;
## conda install ipykernel 

* Sonrasında;
## ipython kernel install –user –name=‘Kayıt_olacak_dosya_ismi’

* Son olarak Jupyter Lab’i terminalde başlatmak için;
## jupyter lab

* Komutları ile işlem yapabiliriz.
* Daha fazla Jupyter komutu için ***Cheatsheet***

# GIT CONFIGURATION

* Git ile işlemlere başlamadan önce hesabımızın konfigürasyonunu yapalım;
## git config –global user.name ‘isminiz’
## git config –global user.email ‘email adresiniz’

* Konfigüre ettiğiniz bilgileri bir metin dosyasında açıp düzenlemek isterseniz;
## git config --global –edit

* Git komutlarının tamamı için Cheatsheet

* Git üzerinde konfigüre ettiğimiz ayarları görmek istersek;
## git config--list 

* Yapılan çalışmamızı local depomuza aktarmak için;
## git init

* Projedeki tüm değişim durumlarını görüntülemek istersek;
## git log

* Kayıt geçmişinin tamamını görüntülemek istersek;
## git status

# Version Control, Git Local, Remote Operations

* Git projemizi Working Directory konumundan git add komutu ile Staging Area konumuna geçirebiliriz. 
## a- git add <dosya_adı> komutu istenilen dosyayı aktarmak için
## b- git add <.> komutu tüm dosyaları aktarmak için kullanılır.

## git add <dosya_adı> or <.> 
* Sonrasında Staging Area konumuna aldığımız çalışmayı local repository üzerine aktarmak istersek;
## git commit -m <iletmek_istenilen_mesaj>

# Gitignore, Branching, Merging

***gitignore***
* Git projeniz içerisinde local repository veya github üzerine eklemek istemediğiniz dosyaları (veri tabanı vb.) .gitignore dosyasına kaydederek git tarafından görmezden gelinmesini sağlayabiliriz. 

***.gitignore dosyası*** içine verimizi ekledikten sonra git add yada git commit komutlarını uygulamalıyız yoksa sonuç alamayız.

* Git ile çalışılan bir proje branch(dallarda)tutulur. Üzerinde farklı bir işlem yapmadığımız taktirde master adlı bir dallanma(branch) üzerinde geliştirmelerimizi yaparız. Branchler git dosyanızın kopyasıdır. Örneğin yazılım projeniz farklı bölümlerden(fronted,backend vb.) oluşuyor. Bu geliştirmeleri farklı branchler üzerinde yaparak hem çalışma karmaşasından uzak hem de  daha sağlıklı bir gelişme ortamı sağlarız. 

* Projemiz üzerine yeni bir branch eklemek istersek;
## git branch <branch_adı>

* Varolan branch üzerinden başka bir branch üstüne geçiş yapmak istersek;
## Git chechout <branch_adı>

* Master branch üzerinde iken branch birleşimi yapmak istersek;
## git merge <branch_adı>

* Projemiz üzerine eklenen branchleri görmek istersek;
## git branch
